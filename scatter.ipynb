```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
df = pd.read_csv('kearlkurve.csv')

c=df[df['Types ']=='Current week']
f=df[df['Types ']=='First week']
plt.figure(figsize=(8,8), dpi=120)
plt.scatter(f['revenue '],f.OI,c='green')
plt.scatter(c['revenue '],c.OI,c='orange')
# df_common = c.loc[c['Branch '].isin(f['Branch '])]




group = df.groupby('Branch ')

for name, df1 in group:
    # plt.quiver(f['revenue '], f["OI"], (df_common["OI"]-df_common['revenue ']),0 )
    plt.plot(df1['revenue '],df1["OI"],c='black')
      
    

for name, df2 in group:
    for i in range(0,1):
        dd = df2[df2.index == df2.index[i]]
        plt.text(dd['revenue '],dd.OI, name, fontdict=dict(color='black', alpha=0.5, size=10))

```
